##### Project Purpose: The iris dataset is an example dataset from the seaborn package. This notebook analyzes the 4 numeric and 1 categorical datatypes from iris through outputs and displays.

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
# Load the dataset from seaborn
df = sns.load_dataset('iris')
print(df.head())
```

```python
# Display 10 rows
print(df.head(10))
```

```python
# Display the number of rows and columns
print(df.shape)
```

```python
# Display the fields and corresponding datatypes
print(df.dtypes)
```

```python
# Display summary statistics for all columns of the dataframe
print(df.describe())
```

```python
# Build historgram for a numerical column
df['petal_width'].hist()
```

```python
#  Build a histogram for each numerical column
df.hist()
```

##### Observation: The sepal length and width is more concentrated in the median values than the petal width and length which is highly distributed towards a 0 petal length and width.

```python
# Count of column 'species'
df['species'].value_counts()

# Inspect value counts for all categorical columns
for col in df.select_dtypes(include=['object', 'category']).columns:
    # Display count plot
    sns.countplot(x=col, data=df)
    plt.title(f'Distribution of {col}')
    plt.show()

# Show all plots
plt.show()
```

##### Observation: There are 150 rows in this dataset. The sepcies category is evenly distributed among setosa, versicolor species, and virginica therefore there are 50 rows of each category.

```python
sns.catplot(
    data = df,
    x='species',
    y='petal_width',
    kind='bar'
)
plt.title('Petal Width by Species Bar Chart')
plt.xlabel('Species')
plt.ylabel('Petal Width')
plt.show()
```

#### Observation: Virginica species has the highest petal width on average.

```python
# Build a violin plot based on species and petal_width
sns.violinplot(x='species', y='petal_width', data=df)
plt.show()
```

#### Observation: Versicolor has less petal width then virginica on average, but more petal width then setosa on average.

```python
# Build a replot ot show relational plots between petal_width and petal_length
sns.relplot(
    data=df,
    x='petal_width',
    y='petal_length',
    hue='species'
)
plt.title('Petal Length vs. Petal Width distribution')
plt.xlabel('Petal Width')
plt.ylabel('Petal Length')
plt.show()
```

#### Among each species, petal length varies little by grouping. For example, versicolor petal length is around 3-5 whereas setosa petal length is around 1-2.
